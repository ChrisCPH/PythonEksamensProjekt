```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def allJobListings():
    yearArr = []
    sumArr = []
    for i in range(2007, 2023, 1):
        data = pd.read_csv('./csv/archiveData/dataByYear/'+str(i)+'kanal.csv')
        dataSum = np.sum(data["Antal"])
        yearArr.append(i)
        sumArr.append(dataSum)
    plt.plot(yearArr, sumArr)
    plt.ylim(0)
    plt.show
plt.figure(figsize=(35,10))
allJobListings() 
```

```python
def areaPercentDistribution(year):
    data = pd.read_csv('./csv/archiveData/percentDataByYear/'+year+'omraadeprocent.csv')
    dist = data['Procent']
    plt.pie(dist, labels=round(data["Procent"], 2))
    plt.legend(data["Navn"])
    plt.show
   
plt.figure(figsize=(35,10))

areaPercentDistribution("2022")
```

```python
def areaPercentData(area):
    data = pd.read_csv('./csv/archiveData/areaPercentData/'+area+' Procent.csv')
    array = np.array(data)
    arr = []
    for i in range (0, len(array), 1):
        appendData = np.array(array[i][2])
        arr = np.append(arr, appendData)
    print(area)
    print('Standard deviation:', arr.std())

areaPercentData("Region Midtjylland")
areaPercentData("Sydjylland")
areaPercentData("Nordsjlland")
areaPercentData("Region Nordjylland")
areaPercentData("Fyn")
areaPercentData("Region Sjlland")
areaPercentData("Bornholm")
areaPercentData("Grnland")
areaPercentData("Skne")
areaPercentData("Udlandet")
areaPercentData("Storkbenhavn")
```

```python
def areaPercentDataPerCategoryPlot(areas):
    for i in range(len(areas)):
        data = pd.read_csv('./csv/archiveData/areaPercentData/'+areas[i]+' Procent.csv')
        plt.plot(data.r, data.Antal)
        plt.legend(areas)
        plt.show

plt.figure(figsize=(35,10))
areaPercentDataPerCategoryPlot(["Nordsjlland", "Storkbenhavn","Region Midtjylland","Sydjylland","Region Nordjylland"])
```

```python
def areaDataPerCategoryPlot(areas):
    for i in range(len(areas)):
        data = pd.read_csv('./csv/archiveData/areaData/'+areas[i]+'.csv')
        plt.plot(data.r, data.Antal)
        plt.legend(areas)
        plt.show

plt.figure(figsize=(35,10))
areaDataPerCategoryPlot( ["Nordsjlland", "Storkbenhavn","Region Midtjylland","Sydjylland","Region Nordjylland"])
```
